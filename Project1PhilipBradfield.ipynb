13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column

```python
from math import sqrt
 
def standard_deviation(list, population):
    num_items = len(list)
    mean = sum(list) / num_items
    diff = [x - mean for x in list]
    sq_diff = [d ** 2 for d in diff]
    ssd = sum(sq_diff)
 
    if population is True:
        var = ssd / num_items
    else:
        var = ssd / (num_items - 1)
    sd = sqrt(var)
    return sd
```

```python
v = standard_deviation(panda_sat_scores['Verbal'])
m = standard_deviation(panda_sat_scores['Math'])
r = standard_deviation(panda_sat_scores['Rate'])

print 'The standard deviation of the average state SAT Verbal scores is ', v
print 'The standard deviation of the average state SAT Math scores is ', m
print 'The standard deviation of the SAT state participation rate is ', r
```

Typically, data from populations studies generate distributions that tend to be "Normal" (Gaussian).  This data
is not looking that way. The rate of participation is bimodal, the Verbal scores are somewhat bimodal as well and the math data is not bell shaped, so something is a little odd.  At this point I am going to take a turn and apply a model to explain what we are seeing.  

What is odd is that in the northeast had much larger rates (close to 90%) than state in the central part of the country like the Dakotas which had rates of 4%.  Being from the mid-west, I did not take the SAT but instead took the ACT - a different college entrance test that is a competitor of the SAT.  A little research on the web turned up a map of the USA that illustrated the states where the ACT is more popular and the states where the SAT is more popular. I used the geographical map to construct a Boolean map of states where the SAT is preferred as compared to states where the ACT is prevelant.   (Even though all of the data here refers to SAT scores, I will refer to the grouping of states where the ACT is more popular as the 'ACT' group and the other group as the 'SAT' group.)  

Applying that Boolean map to the state participation data and plotting the results gives some insight to the data. The previously observed bimodal distribution is broken apart into two groups.  A large majority of those states in the ACT group have a participation rates under 20% and 13 of those 24 states having a participation rate of less than %10).  The distribution of the participation rate SAT group is wider but a wide majority of those states (23 of the 27) have participation rates of more than 50% with 7 of the 27 having participation rates over 70%. 

Repeating the previous procedure on the SAT Verbal score broke the bimodal behaviour into two separate modes and I found this approach insightful on all the data. This report will provide side-by-side graphs that make a fairly strong case that the reason for the non-Normal behaviour is because of the preferance of the ACT test in about half of the states that are generally make up the central part of the nation.
